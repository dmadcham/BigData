# Vectores usando listas
Podemos representar un vector con una lista. Para un vector en el plano, la lista tendrá dos elementos

```python
v1 = [2, 3]
v2 = [4, 5]
```

La suma de dos vectores es la suma componente a componente. Para poder sumar dos vectores, tienen que tener el mismo número de componentes.

```python
v3 = []
v3.append(v1[0] + v2[0])
v3.append(v1[1] + v2[1])
v3
```

Ejemplo: función que recibe una lista y devuelve la de suma todos sus elementos

```python
# recibe una lista y suma todos sus elementos
def sumarVector(lista):
    suma = 0
    for elem in lista:
        suma = suma + elem
    return suma
sumarVector([4, 5, 6])
```

Otros ejemplo para recorrer listas

```python
for elem1, elem2 in zip(v1, v2):
    print(elem1, elem2)
```

```python
for i in range(len(v1)):
    print(v1[i], v2[i])
```

Ejemplo: función que recibe dos listas y devuelve una lista con la suma posición a posición. Si no tienen el mismo número de elementos, lanza una excepción.

```python
def sumaVectores(vec1, vec2):
    resul = []
    for elem1, elem2 in zip(vec1, vec2, strict = True):
        suma = elem1 + elem2
        resul.append(suma)
    return resul
print(sumaVectores([2, 3, 4], [5, 7, 10]))
sumaVectores([2, 3, 4], [5, 7, 10, 4]) # causa excepción
```

Ejemplo: función que recibe dos listas que representan vectores y devuelve su producto escalar. Si no tienen el mismo número de elementos, lanza una excepción.

```python
def productoEscalar(vec1, vec2):
    if(len(vec1) != len(vec2)):
        raise Exception("Error, diferentes longitudes")
    suma = 0
    for elem1, elem2 in zip(vec1, vec2):
        suma = suma + elem1 * elem2

    return suma
print(productoEscalar([2, 3, 4], [5, 7, 10]))
productoEscalar([2, 3, 4], [5, 7, 10, 4]) # causa excepción
```

7. Escribe una función que reciba dos listas de cadenas con el mismo número de elementos. Devolverá una nueva lista con los mismos elementos que la anterior. Cada elemento de la lista será la palabra más larga de los dos argumentos para esa posición. En caso de empate, cogerá la primera por orden alfabético.

```python
def palabrasMasLargas(lista1, lista2):
  resul = []
  for p1, p2 in zip(lista1, lista2, strict = True):
    if( len(p1) > len(p2) or (len(p1) ==  len(p2) and p1 < p2) ):
      resul.append(p1)
    else:
      resul.append(p2)
  return resul
```

```python
a = ['aa', 'aaa', 'aaaa', 'bbbbb']
b = ['b', 'bbbb', 'bbbb', 'aaaaa']
palabrasMasLargas(a, b)
```

8. Escribe una función que reciba una lista y de cadenas y devuelva otra en la que solo estén las tres últimas letras de cada palabra.

```python
def final(palabras):
  finales = [x[-3:] for x in palabras]
  return finales
palabras = ['aaaaa222','nnnnnn111','sfsdlll']
final(palabras)
```

1. Escribe una función que reciba una lista de listas de enteros, y compruebe si es válida como matriz. Es válida si todas las filas tienen el mismo número de columnas.

```python
# si no se le pasa una lista de listas, salta una excepción
def esMatriz(m):
    columnas = len(m[0]) # error si m no tiene len!
    for fila in m:
        if(len(fila)!=columnas):
            return False
    return True
m1 = [[0, 1, 2], [4, 6, 8]]
print(esMatriz(m1))
m1 = [[0, 1, 2], [4, 6]]
print(esMatriz(m1))
```

Recorrer una matriz posición a posición

```python
m1 = [[0, 1, 2], [4, 6, 8]]
for i in range(len(m1)):
    for j in range(len(m1[i])):
        print(m1[i][j], end = " ")
    # para cambiar de fila
    print()
```
