Sucesión Fibonnaci estilo normal (10 primeros términos)

```python
ultimo = 1
penultimo = 0
print(penultimo, end = " ")
print(ultimo, end = " ")
for i in range(8):
    actual = ultimo + penultimo
    print(actual,  end = " ")
    penultimo = ultimo
    ultimo = actual

```

Sucesión Fibonnaci estilo tutorial (10 primeros términos)

```python
# Fibonacci series:
# the sum of two elements defines the next

penultimo, ultimo = 0, 1
print(penultimo, end = " ")
print(ultimo,  end = " ")
for i in range(8):
    actual = ultimo + penultimo
    print(actual,  end = " ")
    penultimo, ultimo = ultimo, actual
```

```python
def procesar_comando(comando):
    match comando:
        # 1. Coincidencia de valor exacto
        case "SALIR":
            print("Cerrando sesión...")

        # 2. Desempaquetado de secuencias (Lista/Tupla) con captura de variables
        case ["ELIMINAR", id_usuario]:
            print(f"Eliminando al usuario con ID: {id_usuario}")

        # 3. Patrón con condicionales (Guards)
        case ["MENSAJE", usuario, texto] if len(texto) < 10:
            print(f"Mensaje corto de {usuario}: {texto}")

        # 4. Captura de múltiples elementos (uso de *)
        case ["NOTIFICAR", *usuarios]:
            print(f"Enviando alerta a {len(usuarios)} usuarios: {usuarios}")

        # 5. Coincidencia con diccionarios (verifica llaves específicas)
        case {"accion": "CONFIGURAR", "valor": v}:
            print(f"Cambiando ajuste a: {v}")

        # 6. Caso por defecto (Wildcard)
        case _:
            print("Comando no reconocido o formato inválido.")

# --- Ejemplos de ejecución ---
procesar_comando(["ELIMINAR", 45])             # Coincide con caso 2
procesar_comando(["MENSAJE", "Ana", "Hola!"]) # Coincide con caso 3
procesar_comando(["NOTIFICAR", "Pedro", "Lucía", "Juan"]) # Coincide con caso 4
procesar_comando({"accion": "CONFIGURAR", "valor": "Modo Oscuro"}) # Coincide con caso 5
```
