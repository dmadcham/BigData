# Pandas - Ejercicio 4

A partir del fichero netflix_titles.csv calcula el número de películas de cada país, su duración máxima, mínima y media.
Guarda esta información en un fichero .csv con con columnas PAIS, NUMERO, MAXIMA, MINIMA Y MEDIA

```python
# Importar librerias
import pandas as pd
import pathlib as Path

# Leer el fichero (desde la carpeta datos)
df = pd.read_csv("datos/netflix_titles.csv")

# Filtrar solo las películas
peliculas = df[df["type"] == "Movie"].copy()

# Limpiar la columna duration
peliculas["duration"] = (
    peliculas["duration"]
    .str.replace(" min", "", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

# Limpiar la columna country (Quitar espacios en balnco y cadenas vacías)
peliculas["country"] = peliculas["country"].str.strip()
peliculas["country"] = peliculas["country"].str.replace('"', '', regex=False)
peliculas["country"] = peliculas["country"].replace("", pd.NA)

# Eliminar filas sin país o sin duración
peliculas = peliculas.dropna(subset=["country", "duration"])

# Separar países cuando hay más de uno (ej: "United States, India")
# Separar por coma (sin asumir espacio)
peliculas["country"] = peliculas["country"].str.split(",")

peliculas = peliculas.explode("country")

# Limpiar espacios y comas sobrantes
peliculas["country"] = peliculas["country"].str.strip()
peliculas["country"] = peliculas["country"].str.rstrip(",")

# Eliminar posibles vacíos generados tras explode
peliculas["country"] = peliculas["country"].replace("", pd.NA)
peliculas = peliculas.dropna(subset=["country"])


# Agrupar por país y calcular estadísticas
resultado = peliculas.groupby("country")["duration"].agg(
    NUMERO="count",
    MAXIMA="max",
    MINIMA="min",
    MEDIA="mean"
).reset_index()

# Renombrar columna country a PAIS
resultado = resultado.rename(columns={"country": "PAIS"})


# Crear carpeta outputs si no existe
Path("outputs").mkdir(parents=True, exist_ok=True)

# Guardar el nuevo CSV en la carpeta outputs
resultado.to_csv("outputs/pelis_ej_pandas_4.csv", index=False)

print("Fichero generado correctamente.")
```

```python
import pandas as pd
from pathlib import Path

# Leer datos
df = pd.read_csv("datos/netflix_titles.csv")

# Filtrar solo películas
peliculas = df[df["type"] == "Movie"].copy()

# Limpiar duración
peliculas["duration"] = (
    peliculas["duration"]
    .str.replace(" min", "", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

# Limpiar country correctamente
peliculas["country"] = (
    peliculas["country"]
    .astype(str)                  # asegurar que es string
    .str.replace('"', '', regex=False)  # quitar comillas
    .str.strip()                  # quitar espacios
    .str.rstrip(",")              # quitar coma final
)

# Eliminar filas inválidas
peliculas = peliculas.dropna(subset=["duration"])
peliculas = peliculas[
    (peliculas["country"] != "") &
    (peliculas["country"].str.lower() != "nan")
]

# Agrupar sin separar países
resultado = (
    peliculas
    .groupby("country")["duration"]
    .agg(NUMERO="count", MAXIMA="max", MINIMA="min", MEDIA="mean")
    .reset_index()
    .rename(columns={"country": "PAIS"})
)


# Crear carpeta outputs si no existe
Path("outputs").mkdir(parents=True, exist_ok=True)

# Guardar CSV
resultado.to_csv("outputs/pelis_ej_pandas_4.csv", index=False)

print("Fichero generado correctamente.")
```
