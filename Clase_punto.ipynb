```python
import math
class Punto:
    def __init__(self, x, y):
        self._x = x
        self._y = y
    def __str__(self):
        return "(" + str(self._x) + "," + str(self._y) + ")"
    def distancia(self, otro):
        return math.sqrt((self._x - otro._x)**2 + (self._y - otro._y)**2)
    def getX(self):
        return self._x
    def getY(self):
        return self._y
    def setX(self, x):
        self._x = x
        return self
    def setY(self, y):
        self._y = y
        return self
```

```python
p1 = Punto(2, 2)
print(p1)
p2 = Punto(3,3)
print (p1.distancia(p2))
```

```python
class Circulo:
    def __init__(self, centro, radio):
        self.centro = centro
        self.radio = radio
    def __str__(self):
        return 'Centro: ' + str(self.centro) + ' Radio: ' + str(self.radio)
    def seTocan(self, otro):
        distancia = self.centro.distancia(otro.centro)
        return distancia <= self.radio + otro.radio
```

```python
p1 = Punto(1,1)
c1 = Circulo(p1, 10)
print(c1)
```

```python
p2 = Punto(1,21)
c2 = Circulo(p2, 9)
print(c2)
```

```python
print(c1.seTocan(c2))
```
