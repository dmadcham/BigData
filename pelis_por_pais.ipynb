```python
import pandas as pd
```

```python
path="../datos/"
datos = pd.read_csv(path + "netflix_titles.csv", sep = ",")
datos.head()
```

```python
# solo pelis
pelis = datos[datos.type=="Movie"].copy()
pelis.duration = pelis.duration.fillna("0 min").copy()
pelis.duration = pelis.duration.str[:-4].astype(int)
agrupado = pelis.groupby("country",as_index=False)["duration"].aggregate(["count","max","min","mean"])
agrupado
```

```python
# alternativa a rename, hay que pasar nuevo nombre para todas las columnas
agrupado.columns = ["PAIS", "NUMERO", "MAXIMO", "MINIMO", "MEDIA"]
agrupado.head()
```

```python
agrupado.to_csv("resumen.csv", index = False)
```
